```python
import pandas  

# Pandas is a library used to analyse data ;Multiple Functions are stored globally  ; When called all the functions in the library will be 
# stored in the memory created for this particular model.
```

```python
import pandas as pd

# "pandas" is called /referred as "pd"
```

```python
dataset=pd.read_csv("Salary_Data.csv")

dataset
#dataset is the name of the variable assigned for data.The dataset file is called using library pd & dot operator 
#Make sure the file is in the same location as the python and specify the name accurately
# If dataset is called we will be able to verify the contents in the file (.csv commaseparatedvariable)
```

```python
independent=dataset[["YearsExperience"]]

independent
# 1 column of the dataset (YearsExperience) is assigned to the variable "independent" (input varibale )
# if multiple columns are needed to assigned separate the columns with commas
# the variable is called to verify if the column data is assigned properly
```

```python
dependent=dataset[["Salary"]]
dependent

# 1 column ( Salary ) of the dataset is assigned to the variable "dependent" (output varibale )
# if multiple columns are needed to assigned separate the columns with commas
# the variable is called to verify if the column data is assigned properly
```

```python
from sklearn.model_selection import train_test_split 

# sklearn is a library with multiple functions ; 
# we are going to define training sets and test sets of data , so we are calling "train_test_split " function from the model_selection module
```

```python
train_test_split(independent,dependent,test_size=0.3,random_state=0)

# we are splitting the input and output variables as training set and test set
# this code returns 60% of input variable and output variables as training set , 30% of input and output variables as test set
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.3,random_state=0)

# the input output train and test sets are assigned names
# X=Input Y=Output
# X_train : input train data ; X_test : input test data
# Y_train : output train data ; Y_test: output test data
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)

# calling the linear_model from the library sklearn and import all the functions in model Linear Regression
# an output variable named regressor is assigned so that all the functions will be saved in the memory of that variable
# calling a function named "fit" and giving the input output training values to train the model; like substituting the values in the formula
#Learning phase happens in the model in this code 
```

```python
weight=regressor.coef_
weight

#LinearRegression() - output like this means the model has been created
#to know values of the weight , we can call using term coef where the values are stored 
```

```python
bias=regressor.intercept_
bias
```

```python
y_pred=regressor.predict(X_test)

#Evaluation : To Evaluate if the model is working properly ; we input the test data and check the regressor score
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
r_score

#calculate the regressor score value by giving test and predicted output data
# if r value <1 the model is efficient and can be saved
```

```python
import pickle
filename="finalised_model_linear.sav"

#pickle is the library used to save files
# Specify the filename and assign it to a variable
```

```python
pickle.dump(regressor,open(filename,'wb'))

#to copy the entire file in pickle we have to use dump command , wb means 'write'
```

```python
loaded_model=pickle.load(open("finalised_model_linear.sav",'rb'))

# Read the loaded model use 'rb'
```

```python
result=loaded_model.predict([[25]])

# Check the results of the loaded model and confirm if it works properly by giving inputs 
```
